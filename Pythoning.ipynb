#### 4. Try and except

The purpose of try and except is that you know that some sequence of instruction(s) may have a problem and you want to add some statements to be executed if an error occurs. These extra statements (the except block) are ignored if there is no error.   
The try and except feature in Pythonis like an ‚Äúinsurance policy‚Äù on a sequence of statements.   
Handling an exception with a try statement is called *catching an exception*. Catching an exception gives you a chance to fix the problem, try again, or at least end the program gracefully

```python
inp = input('Enter Fahrenheit Temperature: ')
try:
  fahr = float(inp)
  cel = (fahr- 32.0) * 5.0 / 9.0
  print(cel)
except:
  print('Please enter a number')
```

#### 5. Short-circuit evaluation of logical expressions

When Python detects that there is noting to be gained by evaluating the rest of a logical expression, it stops the evaluation. When the evaluation of a logical expression
 stops because the overall value is already known, it is called *short-circuiting the evaluation*.   
 The short-circuit behavior leads to a clever
 technique called the guardian pattern.


```python
x = 6
y = 2
x >= 2 and (x/y) > 2
# returns TRUE
```

```python
a = 1
b = 0
a>=2 and (a/b) > 2
# returns FALSE because a>=2 is not fulfilled so it does not go ahead and run the next expression (a/b)>2 due to the short-circuit rule

#p = 4
#q = 0
#p >= 2 and (p/q)>2
# returns ZeroDivisionError - first expression p>=2 is fulfilled so Python starts to evaluate (p/q) when it sees q is 0, hence it gives back error
# the second code was executed with a False result as opposed to the error message in the third code is due to short-circuit behavior
```

#### Defining a function using "def"
Used to define our own functions in the program which we can invoke as needed. We use parantheses to define parameters and indent the body. It only defines the function, does not invoke it.

Eg: def thing():

The *name of the function* is thing.   
The *empty parentheses* after the name indicate that this function doesn‚Äôt take any arguments.   
The first line of the function definition is called the *header*; the rest is called the *body*. The header has to end with a colon and the body has to be indented.   
Defining a function creates a variable with the same name.

```python
def thing():   #function definition
  print("Hello!")
  print("How are you?")
thing()       #function calling/invoking
print("Sam")
thing()       #function calling
print("Kim")
thing()
print("Ray")
```

#### Return values   
Often a function will take its arguments, do some computation, and return a value to be used as the value of the function call in the calling expression.
The keyword "return" is used in this case

```python
#multiple parameters/ arguments

def addtwo(a,b):
  added = a + b
  return added
sums = addtwo(6,5)
print("Sum", sums)

def prod(p,q):
  product = p * q
  return product
product = prod(12,6)
print("Product", product)

print("\n")

#area of a rectangle
len = int(input("Length in:"))
wid = int(input("Width in:"))
area = prod(len,wid)
print("Area of rectangle is", area)
```

#### Loops and its types

**Loops** are repeated steps that have **iteration variables** that change each time through a loop.
   

**Types** -
* *Infinite loops* do not stop by themselves - have to interrupted manually. Infinite loops are created when there is no iteration variable and hence the loop repeats forever.
* *Zero- Trip loops* are constructed in such a way (on purpose often) that the loop never runs for the stated condition. It asks the question, says no, and skips the loop.

#### **Indefinite loop**

Using **while** statement:

One form of iteration in Python is the "while statement".   
Flow of execution for a while statement
  1. Evaluate the conditions, yielding *True* or *False*
  2. If the condition is *False*, exit the while statement and continue execution at the next statement.
  3. If the condition is *True*, execute the body and then go back to step 1.

The body of the loop should change the value of one or more variables (called the *iteration variable*) so that eventually the condition becomes false and the loop terminates.

#### break

Breaking a loop statement

The **break** statement ends the current loop and jumps to the statement immediately following the loop.

In the example below, the loop runs inifinitely till the user inputs 'done', that is when the loop breaks.

```python
while True:
  line = input('> ')
  if line == 'done':
    break
  print(line)
print('Done')
```

#### continue

Finishing iterations with continue

Sometimes in an iteration of a loop we want to finish the current iteration and immediately jump to the next iteration. In that case, we use the **continue** statement to skip to the next iteration without finishing the body of the loop for the current iteration.

In the example below, the code uses "continue" within the loop.  
When line[0] starts with '#' sign followed by 'continue', the statement is not printed, meaning - the rest of the loop is not executed (the loop does not rach print(line)).   
In this case it acts similar to Python comments starting with '#.

```python
while True:
  line = input('> ')
  if line[0] == '#':
    continue
  if line == 'done':
    break
  print(line)
print('done!')
```

#### **Loop Patterns**

Often we use a 'for' or 'while' loop to go through a list of items or the contents of a file and we are looking for something such as the largest or smallest value of the data we scan through. These loops are generally constructed by:
* Initializing one or more variables before the loop starts
* Performing some computation on each item in the loop body, possibly changing the variables in the body of the loop
* Looking at the resulting variables when the loop completes


1. Counting and summing loops
    * Counting - variable count is set to 0 before starting, then it is run through the list of numbers. The iteration variable (named itervar in this example)controls the loop and causes the loop body to be executed once for each of the values in the list. We add 1 to the current value of count for each value in the list.

    * Summing - variable count is set to 0 before starting, then it is run through the list of numbers. Contrary to counting, the iteration variable in used for summing. We add the actual number of the list to the running total during each loop iteration.



2. Maximum and minimum loops

Used to find the largest and the smallest values in a list or sequence. Before the loop the variables (largest and smallest in this example) ar set to the constant 'None' - it marks the variable as empty.

** In practice none of these loops are particularly useful as there ar built-in functions   
len() - counting number of items   
sum() - total of the items   
max() - largest of items   
min() - smallest of items

#### Making **Smart** Loops


* Set the initial value
* Know the entire loop well in your head to try to work something out that runs separately for all the elements where looping will take place
* Check the iteration
* Set a well-defined value to end the loop

```python
count = 0
total = 0
largest = None
smallest = None
print('Largest and smallest before', largest, smallest)
print('Count and total before', count, total)
for itervar in [3, 41, 12, 9, 74, 15]:
  count = count +1
  total = total + itervar
  if largest is None or itervar > largest:        #if statement allows us to filter inside the loop
    largest = itervar
  if smallest is None or itervar < smallest:
    smallest = itervar
print('Count after', count)
print('Total after', total)
print('Largest and smallest after: ', largest, smallest)
```

#### **Search using a Boolean variable**

If we want to search and know if a value was found, we use a variable that starts at False and is set to True as soon as we find what we are looking for.

For the example below, we chose variable named found which is set to False before the loop starts.   
The condition states if value == 3 found will turn True.

The output shows False for values 9 41, and 12; turns True for 3, and **remains True after that**. This is because in the loop, there is no condition that sets found back to False after it has found 3.

```python
found = False     #variable 'found' set at False in the beginning
print('Before', found)
for value in [9, 41, 12, 3, 74, 15]:
  if value == 3:
    found = True    #searching using Boolean -setting variable as True
  print(found, value)
print('After', found)

# Output - False for values 9 41, and 12; turns True for 3, and remains True after that
# because in the loop, there is no condition that sets found back to False after it has found 3.
```

#### Slicing strings

A segment of a string is called a *slice*.

* We can look at any continuous section of a string using the colon operator ':'
* The operator [n : m] returns the part of the string from the n-th character to the m-th character.
* It includes the n-th character in counting but excludes the m-th one.
* If we omit the n-th character [:m], the slice starts at the beginning of the string.
* If we omit the m-th character, [n:], the slice goes to the end of the string.
* If the m-th character is beyond the end of the string, it stops at the end - no trace back error.
* If the first index is greater than or equal to the second the result is an empty string, represented by two quotation marks ' '. An empty string contains no characters and has length 0.

```python
s = 'Monty Python'
index = 0

for st in s:
  print(index, st)
  index += 1

print(s[0:5])         #read as s sub zero through 5
print(s[6:12])        #6:12 does not include 12th char. 6:11 excludes 11th char and does not print the last char
print(s[3:])          #omiting m-th character
print(s[:8])          #omiting n-th character
print(s[0:20])        #m-th character beyond end of string
print(s[:])           #entire string
```

  * It is a method is similar to calling a function (it takes arguments and returns a value) but the syntax is different.
  * We call a method by appending the method name to the variable name using the period as a delimiter. This form of dot notation specifies the name of the method and the name of the string to apply the method to, and the empty parantheses indicate that the method takes no argument.

```python
place = 'kolkata'
print(place)

#method invoking
index = place.find('t')         #invoke find method on place and pass letter t as argument
print(index)
print(place.find('ol'))               #finds and provides index of first letter

#place.find('a', 4)       takes the index as a secoond argument


#capitalise()
print(place.capitalize())
print(place.replace('a', 'A'))
```

* Methods **startswith** and **endswith** returns boolean values. It requires alphabet case to match

```python
#startswith and endswith
var = 'Good morning pineapple'
print(var)

print("1", var.startswith('Good'))       # startswith method - case sensitive

print("2", var.endswith('e'))            # endswith method
print("3", var.endswith('E'))            # endswith method - case senstitive

new_var = var.lower()                    # lower case conversion
print("4", new_var.startswith('g'))      # startswith - not case dependent as string has been converted to lowercase
print("5", new_var.startswith('m'))
```

4. **Parsing and extracting** strings

Often we want to look into a string and find a substring.

For the example below, we want to pull out only the domain of the e-mail address (uct.ac.za) - we can do this using 'find' and string slice.

* Step 1 - Find the position of the *@* sign in the string.
* Step 2 - Find the position of the first space *after* the @ sign
* Step 3 - Use string slicing to extract the position of the string in question

```python
data = "From stephen.marquard@uct.ac.za Sat Jan 5 09:14:16 2008"
print(data)

atpos = data.find('@')                  #finds @ in the data
print(atpos)
sppos = data.find(' ', atpos)           #finds the next space starting at @
print(sppos)

host = data[atpos + 1 : sppos]          #slice string- position starts at atpos and ends at sppos but does not include it
print(host)

fnamepos = data.find('s')
print(fnamepos)
#lnamepos = data.find('d')
lnamepos = data.find('@', fnamepos)               #searches till @ starting from fnamepos
print(lnamepos)
namepos = data [fnamepos : lnamepos]           #since @ is not part of the name, it is excluded per slicing rule, the output is only the name
#namepos = data [fnamepos : lnamepos + 1]      #+1 includes the last character which is part of the name
print(namepos)

name = namepos.replace('.', ' ')        #the . is replaced with a space
print(name.title())                     #the name is capitaized as titlecase

```

#### Formatted string literals

A **formatted string literal** (often referred to simply as an **f-string**) allows Python expressions to be used within string literals. This is accomplished by prefixing an **f** to the string literal and enclosing expressions in curly braces {}.

For example, wrapping a variable name in curly braces inside an f-string will cause it to be replaced by its value. In the example given below, '25' is a string and is not to be confused with integer value 25.

```python
animal = input("Animal: ")
cat = 25
country = input("Country: ")

animal = animal.lower()
print(f"I love {animal}")
print(f'I have {cat - 20} {animal}')              #anything can be embedded as a value
print(f"I spotted {cat} {animal} while travelling in {country.capitalize()}")
```
