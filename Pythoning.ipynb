# Features of Python


1. *Simple and easy to learn* - Python has a simple syntax with only 30+ keywords; has more readability and simplicity; we can also reduce development and project cost
2. *High level programming language* - It is a programmer-friendly language as we are not required to concentrate on activities like memory management, security, etc
3. *Freeware and open source* - The source code of python is open and we can use Python s/w without any license
4. *Platform independent* - Python programs can run on any platform
5. *Portability* - Python programs can be migrated from one platform to another and will provide the same result nonetheless
6. *Dynamically typed* - We are not required to declare type for variables - it provides more flexibility to the programmer
7. *Both procedure and object oriented* - Supports features from both language types
8. *Extensive library* - Has a rich inbuilt library that a programmer can use directly
9. *Interpreted* - We are not required to compile programs explicitly- the internal interpreter takes care of it
10. *Extensible and embedded* - We can use other language programs in Python and Python programs in any other language program!

# Program Steps

* Sequence - steps done in an order
* Conditional - steps skipped at certain places
* Repeated - steps repeated over and over again (computers are super good at this!)

The little code above represents sequential steps - executes in an order.

The code below represents conditional steps - some are skipped based on user input.

```python
x= 25
if x < 10:
  print('smaller')
if x > 20:
  print('bigger')
print('Done!')
```

```python
n = 5
while n > 0:
  print(n)
  n = n - 1
print('Blastoff')
```

This is a repeated steps code with a 'while' loop.

Loops (repeated steps) have iteration variables that change each time through the loop.

As long as n is greater than 0, the intended block will continue to run as repetitive steps.
